# `친밀도전파(Affinity Propagation) 알고리즘을 통해 IRIS 데이터셋을 군집화하기`

```python
from sklearn.cluster import AffinityPropagation

# iris data loading
iris = datasets.load_iris()

X = iris.data
y = iris.target   # 학습을 하는 것이 아닌 군집의 결과가 올바른지 확인하기 위해서

# iris data set 에서 데이터 사이의 거리를 측정하여 최솟값, 메다안, 최댓값을 구하자. (X data 는 총 150 개이다.)

# Similarity Matrix 를 의미하는 리스트를 생성
#sim_matrix = np.array([[0.]*len(X)*len(X)], np.float32).reshape(len(X), len(X))
sim_matrix = np.zeros((len(X),len(X)), np.float32)
for i in range (len(X)):
  main = X[i]
  for j in range(len(X)):
    # sii 는 고려하지 않으므로 그에 따른 분기문 처리
    if j != i:
      value = (-1)*pow(main-X[j],2)
      # 4 차원으로 출력되는 것을 하나로 합쳐준다.
      value = sum(value)
      sim_matrix[i][j] = value

min_value = np.min(sim_matrix)
median_value = np.median(sim_matrix)
max_value = np.max(sim_matrix)
print(f'최솟값 : {min_value} / 중앙값 : {median_value} / 최댓값 : {max_value}')
```

# 1. 친밀도 전파 - 최솟값

```python
# #############################################################################
# 친밀도 전파 알고리즘을 수행한다.

#[1] 자기유사도를 최솟값으로 설정한 경우,
af = AffinityPropagation(preference=min_value).fit(X)

cluster_centers_indices = af.cluster_centers_indices_
labels_affi = af.labels_

print(f'군집의 수 : {len(cluster_centers_indices)}')

# 결과를 시각화해보자
# plotting 을 하기위한 figure
fig_affi = plt.figure(1, figsize=(8, 6))    # (figure 넘버, width_height in inch)
ax_affi = Axes3D(fig_affi, rect=[0,0,0.95,1], elev=0, azim=45)
#ax.scatter(X[:,0], X[:,1], X[:,2], c=kmeans.labels_.astype(float), edgecolor="k")     # 입력의 종류가 4가지 이므로 시각화가 불가능하므로 3개만 가져온다.
                                                                                      # 컬러코드 c 에 대표 인덱스(float) 를 대입하여 색을 부여한다. edgecolor 는 검은색으로 엣지를 표시
ax_affi.scatter(X[:,0], X[:,1], X[:,2], c=labels_affi.astype(float), edgecolor=None)
# 축별로 레이블을 진행하여 축의 의미를 부여한다.
ax_affi.w_xaxis.set_ticklabels([])
ax_affi.w_yaxis.set_ticklabels([])
ax_affi.w_zaxis.set_ticklabels([])   # 축의 좌표자리에 적히는 숫자를 의미한다.
'''
ax.w_zaxis.set_ticklabels(['','A','B','C','D',
                         'E','F','G',''])
'''
ax_affi.set_xlabel("0")
ax_affi.set_ylabel("1")
ax_affi.set_zlabel("2")

print(f'Kmeans Result\n{labels_kmeans_3}\n\nAffinity Propagation Result\n{labels_affi}')   # 개수는 같은데, 당연히 군집화 과정이니 대표번호는 다를 수 있다. 
```
